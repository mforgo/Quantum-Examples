## 1. Introduction
The Phase Gate is a  quantum gate that changes the phase of a qubit. Essentially, we can think about it as it is rotating the vector in Bolchov sphere around its Z-axis. It changes the phase about specified amount, so in some special cases it is the same as other gates changing phase, specificly Z, S and T gates.

The matrix for this gate is
$$
\begin{bmatrix}
1 & 0 \\
0 & e^{iϕ} \\
\end{bmatrix}
$$
where ϕ can be any number from 0 to 2π.

## 2. Special cases
When the ϕ equals some precise numbers, the action performed by this gate is the same as some other gate. Specifically, for ϕ = pi is it the same as Pauli-Z Gate, for ϕ = π/2 is it the same as S-Gate and for ϕ = π/4 is it T-gate.

Because we are rotating the vetor around Z-axis, when state of qubit is exactlly |1> or |0>, nothing happens after apllying this gate.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

qreg_q = QuantumRegister(1, 'q')
creg_c = ClassicalRegister(1, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

plot_bloch_multivector(Statevector(circuit), title="Initial State")
```

```python
from math import *
circuit.p(pi / 5, qreg_q[0])
plot_bloch_multivector(Statevector(circuit), title="After the phase gate")
```

To see the real effect, we have to firstly aplly Hadamard gate.

```python
circuit.h(qreg_q[0])
plot_bloch_multivector(Statevector(circuit), title="Initial state (after Hadamard)")

```

```python
circuit.p(pi / 5, qreg_q[0])
plot_bloch_multivector(Statevector(circuit), title="After Hadamard and Phase Gate (ϕ = pi/5)")   
```

```python
circuit.reset(qreg_q[0])
circuit.h(qreg_q[0])
circuit.p(3*pi / 2, qreg_q[0])
plot_bloch_multivector(Statevector(circuit), title="After Hadamard and Phase Gate (ϕ = 3pi/2)")
```

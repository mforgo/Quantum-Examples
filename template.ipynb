### 2.1 Initialization
Firsly we need that all qubits are in the state $$| - \rangle$$

```python
from qiskit import QuantumCircuit
import math

N = 32  # Number of states, must be a power of 2
n_qubits = int(math.log2(N))
qc = QuantumCircuit(n_qubits)

initial_state = '-' * n_qubits  # Initial state as a string of '0's
qc.initialize(initial_state, range(n_qubits))
```

```python
from qiskit.visualization import plot_histogram

# Example measurement results (counts)
counts = {'00': 10, '01': 5, '10': 2, '11': 33}

# Plot histogram
plot_histogram(counts, title='Measurement Results Histogram')
```

```python
from qiskit.visualization import plot_bloch_vector

# Example Bloch vector (state on Bloch sphere)
bloch_vector = [0.5, 0.5, 0.707] 

# Plot Bloch sphere vector
plot_bloch_vector(bloch_vector, title='Qubit on Bloch Sphere')
```

```python
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
from qiskit import QuantumCircuit

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)

# Get the statevector for the circuit
state = Statevector.from_instruction(qc)

# Plot Bloch spheres for each qubit in the statevector
plot_bloch_multivector(state, title='Multi-Qubit Bloch Spheres')
```
